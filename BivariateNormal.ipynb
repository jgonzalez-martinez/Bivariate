```python
import pandas as pd
import numpy as np
import math as math

rng = np.random.default_rng()


ns = 300
one = np.ones((ns,))#1D list. One element(a list) in the array. ns elements in the list.

data = {
    'mass_1': rng.uniform(100000, 500000, size=(ns,)), 
    'mass_2': rng.uniform(100000, 500000, size=(ns,)),
    'chirp': 1*one, ##how to use the values of keys mass_1 and mass_2?
    'mass_ratio': 1*one,
    'redshift': 0.09 * one,
    'm1-E[m1]':1*one,
    'm2-E[m2]':1*one,
    'luminosity_distance': rng.uniform(0, 500, size=(ns,)),
    'theta_jn': 5/6 * np.pi * one,
    'ra': 3.45 * one,
    'dec': -0.41 * one,
    'psi': 1.6 * one,
    'phase': 0 * one,
    'geocent_time': 1187008882 * one, 
}

data['chirp'] =(((data['mass_1']*data['mass_2'])**(3/5))/(data['mass_1']+data['mass_2'])**(1/5))*one
data['mass_ratio'] =(data['mass_1']/data['mass_2'])*one

##For loop to sum up all the masses 1 and masses 2 
for i in data:
    sum_mass1=+data['mass_1']
    sum_mass2=+data['mass_2']
    sum_mass1sqr=+(data['mass_1']*data['mass_1'])
    sum_mass2sqr=+(data['mass_2']*data['mass_2'])
    sum_mass1mass2=+(data['mass_1']*data['mass_2'])
    
##Computation of E[m1], E[m2], E[m1²], E[m2²].    
exp_mass1=sum_mass1/ns
exp_mass2=sum_mass2/ns
exp_mass1sqr=sum_mass1sqr/ns
exp_mass2sqr=sum_mass2sqr/ns
exp_mass1mass2=sum_mass1mass2/ns

##Terms of interest of the bivariate normal distribution
##sigma1²=E[m1²]-E²[m1]
##sigma2²=E[m2²]-E²[m2]
##sigma12=E[m1m2]-E[m1]E[m2]
data1 = {
    'mass_1': (data['mass_1'])*one, 
    'mass_2': (data['mass_2'])*one, 
    'm1-E[m1]':(data['mass_1']-exp_mass1)*one , 
    'm2-E[m2]':(data['mass_2']-exp_mass2)*one,
    'sigma1sqr': exp_mass1sqr-(exp_mass1**2), 
    'sigma2sqr': exp_mass2sqr-(exp_mass2**2),
    'sigma12': exp_mass1mass2-(exp_mass1*exp_mass2),
    'rho':1*one,
    'z':1*one,
    'p(m1,m2)':1*one,
}

data1['rho']=(data1['sigma12']/(data1['sigma1sqr']*data1['sigma2sqr'])**(1/2))*one

data1['z']=(((data1['m1-E[m1]'])**2)/data1['sigma1sqr']  
+(2*data1['sigma12']*data1['m1-E[m1]']*data1['m2-E[m2]']/((data1['sigma1sqr'])*(data1['sigma2sqr']))) 
+((data1['m2-E[m2]'])**2/data1['sigma2sqr'])*one)

#data1['p(m1,m2)']=((1/(2*np.pi*(data1['sigma1sqr']*data1['sigma2sqr'])**(1/2)))
#*math.exp(-(data1['z']/(2*(1-data1['rho'])**2))))*one

#Computation of sigma1²=E[m1²]-E²[m1]
#data1[sigma1]=
data1 = pd.DataFrame.from_dict(data1)

data1
```
